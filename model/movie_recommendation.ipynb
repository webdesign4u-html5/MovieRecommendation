```python
import numpy as np
import pandas as pd
```

```python
movies= pd.read_csv('data/movies.csv')
credits= pd.read_csv('data/credits.csv')
movies.head(2)
column_index = movies.columns.get_loc('title')
print(column_index)
```

```python
dataset = movies.merge(credits, on='title')
dataset.head(2)
```

```python
dataset = dataset[['movie_id','title','tagline','overview','genres','keywords','cast','crew','popularity','release_date']]
```

```python
dataset.duplicated().sum()
dataset.head(1)
```

```python
dataset['release_date']= dataset['release_date'].apply(lambda x: str(x).split("-")[0] if len(str(x).split("-")) >= 3 else "")
```

```python
dataset.head(2)
```

```python
import ast
def convertCast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=4:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
dataset['cast'] = dataset['cast'].apply(convertCast)
def convertCrew(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
dataset['crew'] = dataset['crew'].apply(convertCrew)
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
dataset['genres']=dataset['genres'].apply(convert)
```

```python
dataset.head(2)
```

```python
popular = dataset[['movie_id','title','genres','release_date','popularity']]
```

```python
popular.head()
```

```python
popular = popular.sort_values('popularity', ascending=False)
```

```python
popular['genres'] = popular['genres'].apply(lambda x: "$".join(x))
```

```python
popular.head()
```

```python
import sqlite3
import sqlalchemy
engine = sqlalchemy.create_engine('sqlite:///dataset.sqlite')
dataset['genres'] = dataset['genres'].apply(lambda x: "$".join(x))
dataset['keywords'] = dataset['keywords'].apply(lambda x: "$".join(x))
dataset['cast'] = dataset['cast'].apply(lambda x: "$".join(x))
dataset['crew'] = dataset['crew'].apply(lambda x: "$".join(x))

dataset.to_sql('dataset', engine)
```

```python
engine = sqlalchemy.create_engine('sqlite:///popular.sqlite')
popular.to_sql('popular', engine)
```
